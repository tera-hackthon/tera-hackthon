```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

sns.set_style('whitegrid')

plt.rcParams['figure.figsize'] = (18.0, 6.0)
```

```python
carac = pd.read_csv("dataset/caracteristics.csv", encoding="latin-1")
```

```python
carac.head()
```

```python
carac.describe()
```

```python
_ = carac.hist()
```

```python
per_year = carac['an'].value_counts().sort_index()
_ = per_year.plot.bar()
```

```python
per_month = carac['mois'].value_counts().sort_index()
_ = per_month.plot.bar()
```

```python
names = {
1:"Full day",
2:"Twilight or dawn",
3:"Night without public lighting",
4:"Night with public lighting not lit",
5:"Night with public lighting on"
}
per_lighting = carac['lum'].value_counts().sort_index().rename(names)
_ = per_lighting.plot.bar()
```

```python
per_dep = carac['dep'].value_counts().sort_index()
_ = per_dep.plot.bar()
# per_dep
```

```python
per_com = carac['com'].value_counts().sort_index()
_ = per_com.plot.bar()
```

```python
names = {
1:"Out of intersection",
2:"Intersection in X",
3:"Intersection in T",
4:"Intersection in Y",
5:"Intersection with more than 4 branches",
6:"Giratory",
7:"Place",
8:"Level crossing",
9:"Other intersection"
}
per_int = carac['int'].value_counts().sort_index().rename(names)
_ = per_int.plot.bar()
```

```python
names = {
1: "Normal",
2: "Light rain",
3: "Heavy rain",
4: "Snow - hail",
5: "Fog - smoke",
6: "Strong wind - storm",
7: "Dazzling weather",
8: "Cloudy weather",
9: "Other"
}
per_atm = carac['atm'].value_counts().sort_index().rename(names)
_ = per_atm.plot.bar()
```

```python
names = {
1:"Two vehicles - frontal",
2:"Two vehicles - from the rear",
3:"Two vehicles - by the side",
4:"Three vehicles and more - in chain",
5:"Three or more vehicles - multiple collisions",
6:"Other collision",
7:"Without collision"
}
per_col = carac['col'].value_counts().sort_index().rename(names)
_ = per_col.plot.bar()
```

```python
names = {
"M":"Métropole",
"A":"Antilles",
"G":"Guyane",
"R":"Réunion",
"Y":"Mayotte"
}
per_gps = carac['gps'].value_counts().sort_index().rename(names)
_ = per_gps.plot.bar()
```
